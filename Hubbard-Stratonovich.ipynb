# Hubbard-Stratonovich via Gauss quadratures

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.special import roots_hermite
```

Get the weights and grid points for [the Gauss–Hermite Formula](https://dlmf.nist.gov/3.5)

```python
x, w = roots_hermite(5)
```

Applying the [the Gauss quadrature](https://dlmf.nist.gov/3.5) to the [Hubbard-Stratonovich_transformation](https://en.wikipedia.org/wiki/Hubbard%E2%80%93Stratonovich_transformation), we obtain
$$
    e^{-\beta y^2} = \frac{1}{\sqrt{\pi}} \int_{-\infty}^{\infty} e^{-x^2} e^{-2i\sqrt{\beta}yx} dx
    \approx \frac{1}{\sqrt{\pi}} \sum_k w_k e^{-2i\sqrt{\beta}yx_k}.
$$

```python
β = 2

λ = 1

y = np.linspace(-5, 8, 200)

HS_approx = sum(
    W * np.exp(- 2j * np.sqrt(β) * (y - λ) * X) for X, W in zip(x, w)
).real / np.sqrt(np.pi)
```

```python
plt.plot(y, np.exp(-β * (y - λ) ** 2), label="exact")
plt.plot(y, HS_approx, "--", label="HS approx")
plt.plot(x, np.exp(-β * (x - λ) ** 2), "o", label="Gauss grids")
plt.legend()
plt.xlabel("$y$")
plt.show()
```
